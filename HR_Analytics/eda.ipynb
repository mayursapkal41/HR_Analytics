```python
#Department-wise Attrition
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

dept_attrition = df.groupby('Department', observed=False)['AttritionFlag'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=dept_attrition.values, y=dept_attrition.index, palette='Reds_r')
plt.title('Attrition Rate by Department')
plt.xlabel('Attrition Rate')
plt.ylabel('Department')
plt.tight_layout()
plt.show()
```

```python
#Salary Band-wise Attrition
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

df['SalaryBand'] = pd.qcut(df['MonthlyIncome'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
salary_attrition = df.groupby('SalaryBand', observed=False)['AttritionFlag'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=salary_attrition.values, y=salary_attrition.index, palette='Blues_r')
plt.title('Attrition Rate by Salary Band')
plt.xlabel('Attrition Rate')
plt.ylabel('Salary Band')
plt.tight_layout()
plt.show()
```

```python
#Attrition vs Years Since Last Promotion
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

promotion_attrition = df.groupby('YearsSinceLastPromotion')['AttritionFlag'].mean().sort_index()

plt.figure(figsize=(8, 5))
sns.lineplot(x=promotion_attrition.index, y=promotion_attrition.values, marker='o')
plt.title('Attrition Rate vs Years Since Last Promotion')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Attrition Rate')
plt.tight_layout()
plt.show()
```
